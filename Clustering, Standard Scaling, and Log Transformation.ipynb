## Now first we apply clustering on this data and verify its importance by comparing with another dataset in which outliers will be removed.

```python
# lets find the appropriate inertia value to find the perfect no. of centroids.

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# lets create a function in which k=2 to 10 will give different value of inertia.

number_clusters = [i for i in range(2,11)]
number_clusters
```

```python
def clustering(number,data_value):
    inertia=[]
    for n in number:
            kmeans = KMeans(n_clusters = n, random_state=42)
            kmeans.fit(data_value)
            inertia.append(kmeans.inertia_)
    return inertia
```

```python
inertias = clustering(number_clusters,dt1)
```

```python
sns.lineplot(x =number_clusters, y =inertias)
```

```python
# lets discover the best k through silhouette_score
def clusteringsil(number,data_value):
    silhouette=[]
    for n in number:
            kmeans = KMeans(n_clusters = n, random_state=42)
            kmeans.fit(data_value)
            silhouette.append(silhouette_score(data_value,kmeans.labels_))
    return silhouette
```

```python
sil_score = clusteringsil(number_clusters,dt1)
```

```python
sns.lineplot(x =number_clusters, y = sil_score)
```

```python
# let say we should take k =3.

kmeans3 = KMeans(n_clusters =3, random_state= 42)
kmeans3.fit(dt1)
```

```python
# lets find out the different attributes
kmeans3.cluster_centers_
```

```python
kmeans3.inertia_
```

```python
silhouette_score(dt1,kmeans3.labels_)
```
